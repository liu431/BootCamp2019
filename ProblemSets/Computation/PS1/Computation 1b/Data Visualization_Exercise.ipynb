```python
import numpy as np
import seaborn as sns
from scipy.special import comb
import matplotlib.pyplot as plt
```

#### Pbm 2

```python
comb(10, 2, exact=False)
```

```python
x = np.linspace(0, 1, 100)
def Bernstein(v, n, x):
    return comb(n, v, exact=False)*(x**v)*(1-x)**(n-v)
```

```python
x = np.linspace(0, 1, 10)
fig = plt.figure(figsize = [10, 10])

plt.subplot(4, 4, 1)
plt.plot(x, Bernstein(0, 0, x))
plt.title("n=0, v=0")

plt.subplot(4,4,5)
plt.plot(x, Bernstein(0, 1, x))
plt.title('n=1, v=0')

plt.subplot(4,4,6)
plt.plot(x, Bernstein(1, 1, x))
plt.title('n=1, v=1')

plt.subplot(4,4,9)
plt.plot(x, Bernstein(0, 2, x))
plt.title('n=2, v=0')

plt.subplot(4,4,10)
plt.plot(x, Bernstein(1, 2, x))
plt.title('n=2, v=1')

plt.subplot(4,4,11)
plt.plot(x, Bernstein(2, 2, x))
plt.title('n=2, v=2')

plt.subplot(4,4,13)
plt.plot(x, Bernstein(0, 3, x))
plt.title('n=3, v=0')

plt.subplot(4,4,14)
plt.plot(x, Bernstein(1, 3, x))
plt.title('n=3, v=1')

plt.subplot(4,4,15)
plt.plot(x, Bernstein(2, 3, x))
plt.title('n=3, v=2')

plt.subplot(4,4,16)
plt.plot(x, Bernstein(3, 3, x))
plt.title('n=3, v=3')

plt.ylim([-0.1,1.1])
plt.show()
```

#### Pbm 3

```python
mlb = np.load('MLB.npy')
```

```python
height = mlb[:,0]
weight = mlb[:,1]
age = mlb[:,2]
```

```python
plt.figure(figsize = [8, 4])
sns.regplot(height, weight)
plt.title('Correlation between height and weight')
plt.xlabel('height')
plt.ylabel('weight')
plt.show()
```

```python
plt.figure(figsize = [8, 4])
sns.regplot(height, age)
plt.title('Correlation between age and height')
plt.xlabel('age')
plt.ylabel('weight')
plt.show()
```

```python
plt.figure(figsize = [8, 4])
sns.regplot(age, weight)
plt.title('Correlation between age and weight')
plt.xlabel('age')
plt.ylabel('weight')
plt.show()
```

#### Pbm 5

```python
X = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(X, X)
Z = (1-X)**2 + 100*(Y-X**2)**2
```

```python
fig, ax = plt.subplots(1,2)
ax[0].pcolormesh(X, Y, Z, cmap='viridis')
ax[1].contour(X, Y, Z, 6, cmap='magma')
plt.plot(1, 1, marker='o' )
plt.show()
```

#### Pbm 6

```python
data = np.load('countries.npy')
```

```python
pop = data[:,0]
gdp = data[:,1]
male = data[:,2]
female = data[:, 3]
```

```python
countries = ["Austria", "Bolivia", "Brazil", "China",
          "Finland", "Germany", "Hungary", "India",
          "Japan", "North Korea", "Montenegro", "Norway",
          "Peru", "South Korea", "Sri Lanka", "Switzerland",
          "Turkey", "United Kingdom", "United States", "Vietnam"]
```

```python
sns.regplot(pop, gdp)
plt.title('Correlation between population and GDP')
plt.xlabel('Population')
plt.ylabel('GDP')
plt.show()
```

```python
sns.regplot(male, female)
plt.title('Male height and female height are correlated')
plt.xlabel('male height')
plt.ylabel('female height')
plt.show()
```

```python
plt.hist(male, bins=10)
plt.title("Histogram of male height")
plt.show()
```

```python
plt.barh(np.arange(20), sorted(gdp), align='center')
plt.yticks(np.arange(20), [countries[i] for i in np.argsort(gdp)])
plt.title("GDP of 20 Countries")
plt.xlabel("in billions $")
plt.show()
```
