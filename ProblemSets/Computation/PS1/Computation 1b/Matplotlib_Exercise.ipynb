```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Pbm 1

```python
def var(x):  
    x = int(x)
    normal = lambda x: np.random.normal(size=(x, x))
    return np.var(np.mean(normal(x), axis=1))
```

```python
plt.ion()
plt.plot(np.linspace(100, 1000, 10), [var(i) for i in np.linspace(100, 1000, 10)])
plt.show()
```

#### Pbm 2

```python
def triplot():
    xvec = np.linspace(-2*np.pi, 2*np.pi, 100)
    plt.plot(xvec, np.sin(xvec), label='sin')
    plt.plot(xvec, np.cos(xvec), label='cos')
    plt.plot(xvec, np.arctan(xvec), label='arctan')
    plt.legend()
    plt.show()
```

```python
triplot()
```

#### Pbm 3

```python
def fnplot():
    x1 = np.linspace(-2,1-1e-10,100)
    plt.plot(x1, 1/(x1-1),  'm--', linewidth = 4)
    x2 = np.linspace(1+1e-10, 6, 100)
    plt.plot(x2, 1/(x2-1),  'm--', linewidth = 4)
    plt.xlim(-2,6)
    plt.ylim(-6,6)
    plt.title(r"$f(x)=\frac{1}{x-1}$")
    plt.plot()
```

```python
fnplot()
```

#### Pbm 4

```python
xvec = np.linspace(0, 2*np.pi, 100)
```

```python
plt.figure(figsize=[8, 4])
ax1 = plt.subplot(221)
ax1.plot(xvec, np.sin(xvec), 'r-')
plt.title('sin(x)')

ax2 = plt.subplot(222)
ax2.plot(xvec, np.sin(2*xvec), 'r--')
plt.title('sin(2x)')

ax3 = plt.subplot(223)
ax3.plot(xvec, 2*np.sin(xvec), 'b--')
plt.title('2*sin(x)')

ax4 = plt.subplot(224)
ax4.plot(xvec, 2*np.sin(2*xvec), 'm:')
plt.title('2sin(2x)')

plt.axis([0, 2*np.pi, -2, 2])
plt.suptitle("Plots")
```

#### Pbm 5

```python
data = np.load('FARS.npy')
hr = data[:,0]
lon = data[:, 1]
lat = data[:, 2]
```

```python
def plots():
    fig, axes = plt.subplots(1,2)
    axes[0].plot(lon, lat, "k,")
    axes[0].set_xlabel('Longitude')
    axes[0].set_ylabel('Latitude')
    axes[0].axis('equal')
    axes[1].hist(hr, bins=int((max(hr)-min(hr))))
    axes[1].set_xlabel('hr') 
```

```python
plots()
```
