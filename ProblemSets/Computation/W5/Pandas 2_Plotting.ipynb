```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
from pydataset import data
```

```python
data()
```

### Data: trees

```python
#data('trees', show_doc=True)
```

```python
df = data('trees')
df.head()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(df.Girth, df.Height, df.Volume, c='r')
plt.title('3D Scatterplot of Trees Data')
ax.set_xlabel('Girth')
ax.set_ylabel('Height')
ax.set_zlabel('Volume')
ax.text(10, 10, 10, 
        r"This data set provides measurements of the girth, height and volume of timber",fontsize=10)
plt.show()
```

```python
df.plot(kind='Hexbin', x='Height', y='Volume', gridsize=10, sharex=False)
ax.set_xlabel('Height')
plt.title('Hexbin plot to show densities of Height and Volume')
plt.show()
```

### Data: Sleep

```python
#data('sleep', show_doc=True)
```

```python
df = data('sleep')
```

```python
df2 = pd.DataFrame({'Group1': df[df['group']==1]['extra'],
                   'Group2': np.array(df[df['group']==2]['extra'])})
df2
```

```python
df2.plot(kind='hist', y=['Group1', 'Group2'], bins=20)
plt.title('Histogram of two groups')
plt.xlabel('extra')
plt.show()
```

```python
df2.plot(kind='box', vert=False)
plt.title('Box plot of the two groups')
plt.xlabel('extra')
plt.show()
```
